```python
import pandas as pd
```

```python
air_data=pd.read_csv('AirQuality_Daily_StudentVersion.csv')

air_data = pd.DataFrame(air_data)
```

# Task 1

```python
air_data_sensorname = air_data.groupby('sensor.name')

air_data_voc = air_data_sensorname.agg(voc_mean =('voc', 'mean'),
                                      voc_median=('voc','median'))
                                      

#air_data_voc
```

```python
top5_voc = (air_data_voc.sort_values(by = ['voc_mean','voc_median'], ascending = False))

top5_voc.head(5)
```

```python
air_data_pm25 = air_data_sensorname.agg(pm25_mean =('pm2.5_atm', 'mean'),
                                      pm25_median=('pm2.5_atm','median'))
#air_data_pm25
```

```python
top5_pm25 = ( air_data_pm25.sort_values(by = ['pm25_mean','pm25_median'], ascending = False))

top5_pm25.head(5)
```

```python
air_data_pm10 = air_data_sensorname.agg(pm10_mean =('pm10.0_atm', 'mean'),
                                      pm10_median=('pm10.0_atm','median'))
#air_data_pm10
```

```python
top5_pm10 = ( air_data_pm10.sort_values(by = ['pm10_mean','pm10_median'], ascending = False))

top5_pm10.head(5)
```

# Task 3

```python
def assign_temperature(temperature):
    if temperature < 32:
        return 'Blow Freezing'
    elif 32<= temperature <=50:
        return 'Cool'
    elif 51 <= temperature <= 70:
        return 'Warm'
    else:
        return 'Hot'
    
air_data['temperature_level'] = air_data['temperature'].apply(assign_temperature)

air_data.head(10)
```

```python
def assign_humidity(humidity):
    if humidity < 50:
        return 'Low Humidity'
    elif 50<= humidity <=80:
        return 'High Humidity'
    else:
        return 'Very High Humidity'
    
air_data['humidity_level'] = air_data['humidity'].apply(assign_humidity)

air_data.head(10)
```

# Task 4

```python
# Unhealthy amounts of PM2.5 for sensitve populations range from 35µg/m³ to 55 µg/m³.
unhealthy_pm25 = air_data[
                            (air_data['pm2.5_atm']>=35) & 
                          (air_data['pm2.5_atm']<=50)
]
    
unhealthy_events = unhealthy_pm25[
    ['date','sensor.name','pm2.5_atm']
].sort_values('date')

unhealthy_events.head(10)
```

```python

unhealthy_pm10 = air_data[
                            (air_data['pm10.0_atm']>=54) & 
                          (air_data['pm10.0_atm']<=154)
]
    
unhealthy_events2 = unhealthy_pm10[
    ['date','sensor.name','pm10.0_atm']
].sort_values('date')

unhealthy_events2.head(10)
```
